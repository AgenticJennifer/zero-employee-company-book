# Voter Propensity — Exploratory Data Analysis

Run `make generate-data` before executing this notebook.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(Path('../data/synthetic/voters.csv'))
print(df.shape)
df.head()
```

```python
# Class balance
print('Positive rate:', df['target_support'].mean().round(3))
df['target_support'].value_counts().plot(kind='bar', title='Support distribution')
```

```python
# Propensity score distribution
df['target_vote_propensity'].hist(bins=50, edgecolor='black')
plt.title('Vote propensity distribution')
plt.xlabel('Propensity score')
plt.ylabel('Count')
```

```python
# County breakdown
df.groupby('county')['target_support'].agg(['count', 'mean']).rename(columns={'mean': 'positive_rate'})
```

```python
# Age distribution by support
df.groupby('target_support')['age'].plot(kind='kde', legend=True)
plt.title('Age distribution by support label')
```
